<h1>ImplicitEquations</h1>

<p>The <code>ImplicitEquations</code> package implements part of an algorithm due to  <a href="http://www.dgp.toronto.edu/people/mooncake/papers/SIGGRAPH2001_Tupper.pdf">Tupper</a> for graphing two-dimensional implicit equations and inequalities.</p>

<p>This file provides some examples. It may be viewed as an <code>IJulia</code> notebook (<a href="http://nbviewer.ipython.org/github/jverzani/ImplicitEquations.jl/blob/master/examples/examples.ipynb">here</a>).</p>

<p>Our first example shows how to graph the <a href="http://www-groups.dcs.st-and.ac.uk/~history/Curves/Devils.html">Devils curve</a> over the default region as follows:</p>

```julia
using PyPlot
pygui(false)
figure(figsize=(5.12, 3.84))
using ImplicitEquations


a,b = -1,2
f(x,y) = y^4 - x^4 + a*y^2 + b*x^2
r = (f==0)
plot(r)
```

<p>The <code>f == 0</code> expression above creates a <code>Predicate</code> that is graphed by <code>plot</code>. For all but the case of <code>f==g</code> (for two functions) <code>Predicate</code>s are generated by overloading the logical operators for functions on the left and real values or functions on the right: <code><</code>, <code><=</code> (<code>\le[tab]</code>), <code>\Equal[tab]</code>, <code>\lessgtr[tab]</code>, <code>>=</code> (<code>\ge[tab]</code>), and <code>></code> . As <code>==</code> and <code>!=</code> can already be used to compare functions in a different way, we use either <code>eq(f,g)</code> or <code>neq(f,g)</code> for these comparisons. Should infix notation be preferred, the Unicode operator <code>f \Equal[tab] g</code> or <code>f \lessgtr[tab] g</code> may be used.</p>

<p>For example, the <a href="http://www-history.mcs.st-and.ac.uk/Curves/Trident.html">Trident of Newton</a> can be represented in Cartesian form as follows:</p>

```julia
## trident of Newton
c,d,e,h = 1,1,1,1
f(x,y) = x*y
g(x,y) =c*x^3 + d*x^2 + e*x + h
plot(eq(f,g)) ## aka f â©µ g (using Unicode\Equal[tab])
```

<p>Inequalities can be graphed as well</p>

```julia
f(x,y)= (y-5)*cos(4*sqrt((x-4)^2 + y^2))
g(x,y) = x*sin(2*sqrt(x^2 + y^2))
r = f < g
plot(r, -10, 10, -10, 10, W=2^9, H=2^9, show_red=true)  # xmin, xmax, ymin, ymax, pixels_wide, pixels_high
```

<p>The coloring scheme employed (mostly) follows Tupper and is based on pixels:</p>

<p>- if a pixel is white, there are no solutions of the predicate <code>r</code> within the pixel;</p>

<p>- if a pixel is black, there is at least one solution of the predicate <code>r</code> within the pixel;</p>

<p>- if a pixel is red, there may or may not be solutions of the predicate <code>r</code> within the pixel. (However, we check these pixels probabilistically and color them black or white.) Red pixels show when the keyword argument <code>show_red</code> is assigned to be <code>true</code>, as above.</p>

<p>A "pixel" represents a region comprised of two intervals. The excellent <code>ValidatedNumerics</code> package is used for the underlying interval arithmetic.</p>

<p>(In the above pixel does not refer to screen pixels for all the graphical outputs, only for a few.)</p>

<p>This static figure shows the algorithm:</p>

<p><img src="http://i.imgur.com/8Mtmb7v.png" alt="Algorithm"></img></p>

<p>The basic algorithm is to initially break up the graphing region into square regions. (This uses the number of pixels, which are specified by <code>W</code> and <code>H</code> above.)  There regions are checked for the predicate. If definitely not, the region is painted white; if definitely yes, the region is painted black; else the square region is subdivided into 4 smaller regions and the above is repeated until subdivision would be below the pixel level. At which point, the remaining 1-by-1 pixels are checked for possible solutions, for example for equalities where continuity is known a random sample of points is investigated with the intermediate value theorem.</p>

<p>This example, the <a href="http://yangkidudel.wordpress.com/2011/08/02/love-and-mathematics/">Batman equation</a>, uses a few new things: the <code>screen</code> function is used to restrict ranges and logical operators to combine predicates.</p>

```julia
f0(x,y) = ((x/7)^2 + (y/3)^2 - 1)  *   screen(abs(x)>3) * screen(y > -3*sqrt(33)/7) 
f1(x,y) = ( abs(x/2)-(3 * sqrt(33)-7) * x^2/112 -3 +sqrt(1-(abs((abs(x)-2))-1)^2)-y)
f2(x,y) = y - (9 - 8*abs(x))       *   screen((abs(x)>= 3/4) &  (abs(x) <= 1) )
f3(x,y) = y - (3*abs(x) + 3/4)     *   I_((1/2 < abs(x)) & (abs(x) < 3/4))    # alternate name for screen
f4(x,y) = y - 2.25                 *   I_(abs(x) <= 1/2) 
f5(x,y) = (6 * sqrt(10)/7 + (1.5-.5 * abs(x)) - 6 * sqrt(10)/14 * sqrt(4-(abs(x)-1)^2) -y) * screen(abs(x) >= 1)

r = (f0==0) | (f1==0) | (f2== 0) | (f3==0) | (f4==0) | (f5==0)
plot(r, -7, 7, -4, 4)
```

<p>- predicates can be joined logically with <code>&</code>, <code>|</code>. Use <code>!</code> for negation.</p>

<p>- The <code>screen</code> function can be used to restrict values according to   some predicate call.</p>

<p>- the logical comparisons such as <code>(abs(x)>= 3/4) & (abs(x) <= 1)</code>   within <code>screen</code> are not typical in that one can't write   <code>3/4 <= abs(x) <= 1</code>, a convenient <code>Julian</code> syntax. This is due to the fact that the "<code>x</code>s"   being evaluated are not numbers, rather intervals via   <code>ValidatedNumerics</code>. For intervals, values may be true, false or   "maybe" so a different interpretation of the logical operators is   given that doesn't lend itself to the more convenient notation.</p>

<p>- rendering can be slow. There are two reasons: images that require a   lot of checking, such as the inequality above, are slow just because   more regions must be analyzed. As well, some operations are slow,   such as division, as adjustments for discontinuities are slow. (And   by slow, it can means really slow. The difference between rendering   <code>(1-x^2)*(2-y^2)</code> and <code>csc(1-x^2)*cot(2-y^2)</code> can be 10 times.)</p>

<h2>A "typical" application</h2>

<p>A common calculus problem is to find the tangent line using implicit differentiation. The difficulty here is adding a layer to a graph produced by <code>plot(r::Predicate)</code>. How to do this varies with the underlying plotting package. Rather than cook up package-specific solutions, we can just use <code>|</code>:</p>

```julia
f(x,y) = x^2 + y^2
a,b = 3,3
dydx(a,b) = -b/a             # implicit differentiate to get dy/dx =-y/x
tl(x) = b + dydx(a,b)*(x-a)

g(x,y) = y - tl(x)

r = (f == f(a,b)) | (g == 0)
plot(r)
```

<h2>Display</h2>

<p>The graphs can be rendered in different ways.</p>

<ul><li>The `asciigraph` function is always available and makes a simple text-based plot.</li></ul>

<ul><li>When one of several plotting packages are loaded, their `plot` function is extended to display a graph when the first argument is a `Predicate`, as generated by the expressions above. For each package a separate plotting function is exported:</li></ul>

<p>- for <code>Winston</code>, the function  <code>wgraph</code> does the work,</p>

<p>- for <code>PyPlot</code>, the function  <code>pgraph</code> does the work, and</p>

<p>- for <code>Gadfly</code>, the function  <code>ggraph</code> does the work. (<code>Gadfly</code> must be loaded prior to <code>ImplicitEquations</code>.)</p>

<ul><li>For rendering with `Cairo` and `Gtk`, a function `cgraph` is exported, but no `plot` method is introduced (not quite right)</li></ul>

<p>This packages use the <code>Requires.jl</code> package to conditionally load modules. In theory the plotting packages can be loaded before or after <code>ImplicitEquations</code>, though in practice that can raise errors. arise.</p>

<h2>Alternatives</h2>

<p>Many such plots are simply a single level of a contour plot. Contour plots can be drawn with the help of the <code>Contour.jl</code> package. They are implemented directly in <code>Gadfly</code> and <code>PyPlot</code>.</p>

<p>The <code>SymPy</code> package exposes SymPy's <code>plot_implicit</code> feature that will implicitly plot a symbolic expression in 2 variables including inequalities. The algorithm there also follows Tupper and uses interval arithmetic, as possible.</p>
